```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
q0 = 2e6;
t0 = 520;
t = 530;
p0 = 14.7;
p2 = 21.7;
rho = 0.7;
eta = 0.7;

func = lambda l,d: np.sqrt(p2**2 + ((l*rho*t)/d**(5.334) * eta**2) * ((q0*p0) / ((433.54*t0))**2))

L = np.linspace(0.1, 0.2)
D = np.linspace(2, 5)

D, L = np.meshgrid(D, L)
P1 = func(L, D)

fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')
# ax.set_zlim(0, 25)
surf = ax.plot_surface(D,L,P1)
ax.set_xlabel('Diameter')
ax.set_ylabel('Length')
ax.set_zlabel('P1')
plt.savefig('plot.png')
```
